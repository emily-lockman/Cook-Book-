# **1.6 Defining and Implementing a Function**

```python
def salary(AnnualSalary):
    min_salary = df_xlsx["Annual Salary"].min()
    max_salary = df_xlsx["Annual Salary"].max()
    avg_salary = df_xlsx["Annual Salary"].mean()
    
    if AnnualSalary < min_salary or AnnualSalary > max_salary:
        return "NaN"
    elif AnnualSalary < avg_salary:
        return "Low Salary"
    else:
        return "Average Salary" if AnnualSalary < 2*avg_salary else "High Salary"
```

This function, `salary(AnnualSalary)`, takes an **AnnualSalary** as an argument, which is a numeric value representing the annual salary of an employee. The function calculates the minimum salary, maximum salary, and average salary of all employees in the dataframe `df_xlsx` using the `min()`, `max()`, and `mean()` functions, respectively. <br> <br>

The function then checks **if** the given AnnualSalary is outside the range of minimum and maximum salaries. If so, it returns <font color = red>"NaN" </font>to indicate that the salary is not valid. <br> <br>

The next **elif statement**checks if the given AnnualSalary is less than the average salary. If so, it returns  <font color = red>"Low Salary" </font>to indicate that the employee has a low salary. <br> <br>

The final **else statement** checks if the given AnnualSalary is greater than or equal to the average salary. If so, it checks if the annual salary is less than twice the average salary. If so, it returns <font color = red>"Average Salary"</font> to indicate that the employee has an average salary. Otherwise, it returns <font color = red>"High Salary"</font> to indicate that the employee has a high salary. <br> <br>

**Implementing Function and Adding a column to store results**

The new column is named <font color = red>"Salary Category"</font>, and its values are computed using a function called salary. Specifically, the `.apply()` method is used to apply the salary function to each value in the "Annual Salary" column of the DataFrame.

```python
df_xlsx["Salary Category"] = df_xlsx["Annual Salary"].apply(salary)
df_xlsx
```
